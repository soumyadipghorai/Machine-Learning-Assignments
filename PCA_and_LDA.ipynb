# Import

```python
import pandas as pd 
import numpy as np
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
dataset = load_digits()
dataset.keys()
```

```python
dataset.data.shape
```

```python
dataset.data[0]
```

```python
dataset.data[0].reshape(8,8) # convert into 2d array
```

# EDA

```python
plt.gray()
for i in range(10) : 
    plt.matshow(dataset.data[i].reshape(8,8)) # ploting a few numbers 
```

```python
np.unique(dataset.target) # unique numbers in target
```

```python
df = pd.DataFrame(dataset.data, columns= dataset.feature_names)
df.head() # creating a dataframe
```

```python
df.describe()
```

```python
x = df
y = dataset.target
y
```

```python
scaler = StandardScaler()

x_scaled = scaler.fit_transform(x)
x_scaled
```

# Fit model

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size= 0.2, random_state= 30)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
model.score(x_test, y_test) # accuracy
```

```python
x.shape
```

# PCA 

```python
pca = PCA(0.95)# retain 95% of the useful features 

x_pca = pca.fit_transform(x)
x_pca.shape
```

```python
pca.explained_variance_ratio_ # column wise variance captured
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size= 0.2, random_state= 30)

model = LogisticRegression(max_iter = 1000) # for gradient descent to converge 

model.fit(x_train, y_train)
model.score(x_test, y_test) # almost same accuracy
```

# PCA with n_components

```python
pca = PCA(n_components= 2) # how many componenst you want
x_pca = pca.fit_transform(x)
x_pca.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_pca, y, test_size= 0.2, random_state= 30)

model = LogisticRegression(max_iter = 1000) # for gradient descent to converge 

model.fit(x_train, y_train)
model.score(x_test, y_test) # accuracy drops drustically
```

# LDA with n_components

```python
lda = LinearDiscriminantAnalysis(n_components = 2)

x_lda = lda.fit(x,y).transform(x)
x_lda.shape
```

```python
lda.explained_variance_ratio_
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_lda, y, test_size= 0.2, random_state= 30)

model = LogisticRegression(max_iter = 1000) # for gradient descent to converge 

model.fit(x_train, y_train)
model.score(x_test, y_test) # accuracy drops drustically
```

So with n_components LDA performs better than PCA here
